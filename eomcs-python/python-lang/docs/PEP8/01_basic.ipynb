# 파이썬 코드 스타일 가이드(Style Guide for Python Code)

이 문서는 파이썬 코드의 스타일 가이드인 PEP 8(Python Enhancement Proposal 8 - Style Guide for Python Code)을 요약한 것이다.
또한 이해에 도움을 주기 위해 예제 코드를 추가하였다.

원문은 [https://peps.python.org/pep-0008/](https://peps.python.org/pep-0008/) 에서 확인할 수 있다.

**코딩 스타일 가이드가 필요한 이유**는 다음과 같다.
- 코드를 작성하는데 사용하는 시간 보다 **코드를 읽고 이해하는데 더 많은 시간을 사용**한다.
- 따라서 **코드의 가독성을 높이는 것**은 개발 생산성을 높이는 중요한 요소이다.
- 코드의 가독성을 높이려면 **일관된 코딩 스타일을 유지**하는 것이 중요하다.

프로젝트에서 사용하는 자체 코딩 스타일 가이드가 있다면 그 가이드를 따르는 것이 좋다.
PEP 8의 규칙을 준수하기 위해 이전 버전과의 호환성을 깨지 말라!


## 코드 레이아웃(Code Lay-out)

### 들여쓰기(Indentation)

기본으로 들여쓰기는 레벨 당 4개의 공백을 사용하라.


**함수를 정의할 때 파라미터를 들여쓰기 하는 방식:**

```python
# PEP8 권장 방식
# - 파라미터와 본문을 구분하기 위해 함수 이름을 기준으로 4칸 들여쓰기를 한다.
# - 하지만 Black, autopep8, YAPF 같은 자동 포맷터와 충동하는 경우가 많다.
def long_function_name(
        var_one, var_two, var_three,
        var_four):
    print(var_one)

# 실무에서 권장하는 방식
# - 자동 포맷터(예: Black, autopep8, YAPF)가 사용하는 스타일이다.
# - 읽기 쉽고, 추가 들여쓰기가 없어 Git diff가 덜 지저분하다.
# - 대규모 프로젝트(Django, FastAPI, Pandas 등)에서 널리 사용한다.
# - diff 충돌을 최소화하고 아규먼트 추가를 쉽게하기 위해 마지막 아규먼트 뒤에도 콤마를 붙인다.(trailing comma)
def long_function_name(
    var_one, 
    var_two, 
    var_three,
    var_four, # 마지막 아규먼트 뒤에도 콤마를 붙인다
):
    print(var_one)
```

**함수를 호출할 때 아규먼트를 들여쓰기 하는 방식:**

```python
# Opening parenthesis 정렬(거의 사용하지 않음)
# - 첫 번째 인수 위치에 맞춰서 나머지를 수직 정렬한다.
# - 함수명 길이에 따라 들여쓰기가 달라져서 읽기 어렵다.
# - 자동 포맷터(예: Black, autopep8, YAPF)에서는 이 스타일을 자동 제거한다.
foo = long_function_name(100, 200,
                         300, 400)

# Hanging Indentations(PEP8 권장 방식)
# - 첫 줄에 아규먼트를 배치하지 않는다.
# - 다음 줄에 들여쓰기 레벨을 하나 더 올린다.
foo = long_function_name(
    100, 200,
    300, 400)

# 실무에서 권장하는 방식
# - 자동 포맷터 Black의 기본 스타일이다.
# - diff 충돌을 최소화하고 아규먼트 추가를 쉽게하기 위해 마지막 아규먼트 뒤에도 콤마를 붙인다.(trailing comma)
foo = long_function_name(
    100, 
    200,
    300, 
    400, # 마지막 아규먼트 뒤에도 콤마를 붙인다.
)

# 짧은 경우 한 줄로 작성
foo = long_function_name(100, 200, 300, 400)
```

**if/while/for 들여쓰기**

```python
# 조건이 적은 경우 괄호없이 한 줄로 작성
if condition_one and condition_two and condition_three:
    do_something()

# 백슬래시(\)를 사용하여 여러 줄로 나눌 수 있다.(과거 스타일, 권장하지 않음)
if condition_one and \
   condition_two and \
   condition_three:
    do_something()

# PEP8 방식
# - 추가 들여쓰기를 하지 않음(No extra indent)
if (this_is_one_thing and
    that_is_another_thing):
    do_something()

# - 가독성을 위해 조건과 내부 블록 사이에 주석을 삽입하기도 한다.
if (this_is_one_thing and
    that_is_another_thing):
    # 복잡한 조건을 설명하는 주석
    do_something()

# - 이어지는 조건에 들여쓰기 레벨을 추가하기(과거 스타일, 권장하지 않음)
if (this_is_one_thing
        and that_is_another_thing):
    do_something()

# 실무에서 권장하는 방식
# - 조건 줄과 내부 블록이 동일한 들여쓰기 레벨을 갖는다.
# - 가독성, 안전성, 자동 포맷터 호환성 모두에 좋은 스타일이다.
if (
    this_is_one_thing and
    that_is_another_thing
):
    do_something()
```

**닫는 괄호(예: ), ], }) 위치**

```python
# 리스트의 마지막 요소 아래에 닫는 대괄호(])를 배치하는 방식
my_list = [
    1, 2, 3,
    4, 5, 6,
    ]
result = some_function_that_takes_arguments(
    'a', 'b', 'c',
    'd', 'e', 'f',
    )

# 실무에서 권장하는 방식
# - 닫는 괄호를 마지막 요소와 같은 들여쓰기 레벨에 배치한다.
# - 자동 포맷터(예: Black, autopep8, YAPF)가 사용하는 스타일이다.
# - trailing comma를 붙여서 diff 충돌을 최소화한다.
my_list = [
    1, 2, 3,
    4, 5, 6,
]
result = some_function_that_takes_arguments(
    'a', 'b', 'c',
    'd', 'e', 'f',
)

# - 자동 포맷터 Black의 기본 스타일
my_list = [
    1,
    2,
    3,
    4,
    5,
    6,
]
```

### 줄 최대 길이(Maximum Line Length)

- **모든 줄은 79자**를 넘지 않도록 한다. 단, **팀 합의가 있다면 99자**까지 허용한다.
- **docstring**이나 **주석**은 **72자**를 넘지 않도록 하라.
- 자동 줄바꿈은 사용하지 않는다.

**실무 방식:**
- Black(포매터), Ruff(포맷터 + 린터): **88자**
- Google Python Style Guide: 80자 or 100자
- Django, FastAPI, NumPy 등 많은 오픈소스: 88~100자 범위
- docstring이나 주석은 72자 권장(PEP 8 그대로 유지)

**편집기 창 너비를 제한하면,**
- 여러 파일을 나란히 놓고 비교하기 쉽다.
- 코드 리뷰 도구를 사용하여 두 버전의 코드를 비교할 때 효과적이다.

**줄을 여러 줄로 나누는 방법:**

- 암시적 줄 이어짐 기능 사용(implied line continuation)

```python
# - 괄호 `( )`, 대괄호 `[ ]`, 중괄호 `{ }` 안에서 줄을 나눈다. 그러면 파이썬이 자동으로 이어준다.
result = (
    some_function(arg1, arg2, arg3)
    + another_function(long_argument_name)
)
```

- 백슬래시(\) 사용 (과거 스타일, 권장하지 않음)

```python
# Black/Ruff 자동 포매터 사용할 때 대부분 자동으로 괄호 스타일로 바뀜
# - 마지막에 공백이 있으면 문법 오류 가능
result = some_function(arg1, arg2, arg3) \
    + another_function(long_argument_name)

# - 들여쓰기가 애매해짐
result = some_function(arg1, arg2, arg3) \
         + another_function(long_argument_name)
```

- 예외적인 백슬래시 사용 1: Python 3.10 이전의 다중 with 문

```python
with open('a.txt') as f1, \
     open('b.txt', 'w') as f2:
    f2.write(f1.read())
```

- 예외적인 백슬래시 사용 2: assert 문

```python
assert user.is_active \
    and user.age >= 18 \
    and user.email.endswith("@example.com"), \
    "User validation failed: inactive or invalid profile"

# 실무에서 권장하는 방식
assert (
    user.is_active
    and user.age >= 18
    and user.email.endswith("@example.com")
), "User validation failed: inactive or invalid profile"
```

- 구문구조가 잘 드러나도록 **들여쓰기**를 한다. 

```python
long_function_call(
    first_argument,
    second_argument,
    another_long_argument,
)

# PEP8 원칙 기반 수동 포맷팅
query = (
    User.objects
    .filter(is_active=True, email__endswith="@example.com")
    .first()
)

# Black/Ruff 자동 포맷팅
query = User.objects.filter(
    is_active=True,
    email__endswith="@example.com",
).first()
```

### 줄바꿈(Should a Line Break Before or After a Binary Operator?)

- 수식이 여러 줄로 나뉠 때는 **이항 연산자 앞에서 줄바꿈** 한다.

```python
# 옛날 스타일(권장하지 않음): 가독성을 떨어뜨림
# - 연산자가 같은 열에 정렬되지 않음
# - 앞에 있는 피연산자와 멀어짐
# - 어떤 연산이 더해지고 빠지는지 한눈에 보이지 않음
income = (gross_wages +
          taxable_interest +
          (dividends - qualified_dividends) -
          ira_deduction -
          student_loan_interest)

# 권장 스타일: 수학적 표기 관습(Donald Knuth의 Typesetting 규칙)
# - 연산자 앞에서 줄바꿈
# - 모든 연산자가 같은 열에 정렬됨 => 한눈에 더하고 빼는 항목이 잘 보임
# - 연산자와 그 피연산자가 시각적으로 함께 묶임
# - 긴 수식이라도 수캔하기가 쉬움
income = (gross_wages
          + taxable_interest
          + (dividends - qualified_dividends)
          - ira_deduction
          - student_loan_interest)
```

### 빈 줄(Blank Lines)

- 최상위 함수와 클래스 정의 사이에는 빈 줄 2개로 둘러싼다.

```python
    # 빈 줄
def foo():
    pass
    # 빈 줄
    # 빈 줄
class Bar:
    pass
    # 빈 줄
```

- 클래스 내부 메서드는 빈 줄 1개로 구분한다.

```python
class User:
    def __init__(self):
        self.name = "Alice"
      # 빈 줄(메서드 사이에 두 줄을 두지 않는다. 코드 길이가 과도하게 늘어남)
    def greet(self):
        print("Hello")
```

- 관련된 함수끼리는 추가 빈 줄로 묶을 수 있다.

```python
    # 빈 줄
def load_config():
    ...

def parse_config():
    ...
    # 빈 줄
    # 빈 줄(메서드 그룹을 구분할 수 있게 적당히 빈 줄을 추가할 수 있다.)
def connect():
    ...

def disconnect():
    ...
    # 빈 줄
```

- 한 줄짜리 함수는 빈 줄 없이 묶어도 된다.

```python
def do_nothing(): pass
def noop(): pass
def dummy(): pass
```

- 함수 내에서 논리적으로 구분되는 코드 블록 사이에는 빈 줄 1개를 넣을 수 있다.

```python
def process_user(user):
    if not user.is_active:
        raise ValueError("Inactive")
        # 빈 줄(이런 식으로 빈 줄을 넣을 수 있다.)
    normalized = user.name.lower()
        # 빈 줄(단 빈 줄이 너무 많으면 흐름이 끊어져 코드를 읽기 힘들다.)
    save_user(normalized)
```

- 파이썬은 페이지 구분 용도로 Control-L 문자(form feed, `\f`)를 허용한다. (실무에서는 거의 사용하지 않는다.)

### 소스 파일 인코딩(Source File Encoding)

- 파이썬 소스 파일(.py)은 **UTF-8**로만 작성해야 한다.
    - 파이썬 3.x 버전부터 `# -*- coding: utf-8 -*-` 선언을 추가할 필요가 없다.

```python
# ❌ 금지: UTF-8 선언을 추가하지 말라.
# -*- coding: utf-8 -*-
def foo():
    ...

# ✔ 권장: UTF-8을 선언할 필요가 없다.
def foo():
    ...
```

- non-ASCII 문자(한글, 중국어, 악센트 문자 등)
    - docstring, 주석, 에러메시지, 고정 문자열, 테스트 문자열
        - **장소**나 **사람 이름**을 나타낼 때만 사용하는 것이 좋다.
    - 문자열 값 안에서 사용하는 것은 괜찮다.
        - 단, 인코딩 오인식, diff 충돌, 코드 리뷰 오류 등을 발생시킬 수 있는 문자는 사용하지 말아야 한다.
            - 시각적 잡음의 유니코드 문자(예: `z̯̯͡a̧͎̺l̡͓̫g̹̲o̡̼̘`)
            - 바이트 순서 표시(BOM; Byte Order Mark) 같은 특수 문자

```python
# 문자열 데이터에는 한글 등 non-ASCII 문자를 사용할 수 있다.
message = "안녕하세요"

# ❌ 금지: noisy 유니코드 문자는 사람이 읽기 어렵고 diff/merge 문제가 발생한다.
text = "h̷e̴l̶l̸o̶"
```

- 식별자(identifier)에는 ASCII 문자만 사용한다.
    - 클래스/함수/변수 이름에 ASCII 이외의 문자를 사용하지 않는다.

```python
# ❌ 금지: ASCII 이외의 문자를 식별자에 사용하지 말라.
사용자이름 = "Tom"

# ✔ 권장: 협업과 언어적 혼란을 피하기 위해 식별자는 ASCII 문자만 사용하라.
user_name = "Tom"
```

    - `학생관리.py` (X)

### 모듈 레벨 던더 이름(Module Level Dunder Names)

- "던더(dunder)" == double underscore
    - 예) `__name__`, `__file__`, `__doc__`, `__all__` 등

- `__init__.py` 파일과 `__all__` 던더 이름
    - `__init__.py` 파일은 패키지 공개 인터페이스를 만들어내는 파일이다.
        - 패키지 초기화
        - 서브 모듈 import
        - public API를 구성
    - 따라서 `__init__.py` 파일을 import 한 후에 모듈 레벨 던더 이름을 정의하는 것은 적절하다.

```python
# 패키지 디렉토리 구조
"""
mypackage/
    __init__.py
    models.py
    service.py
"""
```

```python
# models.py
class User:
    def __init__(self, name: str):
        self.name = name

    def greet(self):
        return f"Hello, {self.name}!"
```

```python
# service.py
from .models import User

class UserService:
    def register(self, name: str) -> User:
        print(f"Registering user: {name}")
        return User(name)
```

```python
# __init__.py
"""
mypackage: A simple example package.
"""

# 패키지 public API로 노출할 클래스 import
from .models import User
from .service import UserService

# 공개할 이름 정의 (외부에서 mypackage.* 로 접근 가능)
# - 패키지 public API 관리: 필요없는 내부 모듈/이름이 외부로 노출되지 않도록 한다.
# - IDE 자동완성/문서화 개선: 공개 API만 목록에 나타난다.
# - `from mypackage import *` 보호: 꼭 필요한 것만 외부로 export 한다.
# - 패키지 인터페이스를 단순화 한다.
__all__ = ["User", "UserService"]
```

```python
# 패키지 사용 예:

# __all__을 사용하지 않을 경우, 내부 구조가 모두 노출됨
# from mypackage.models import User

# __all__을 사용한 경우, 공개된 클래스만 import 가능
from mypackage import User, UserService

service = UserService()
user = service.register("Alice")

print(user.greet())
```

## 문자열 따옴표(String Quotes) 

- 단일 인용부호 ' ' 와 이중 인용부호 " " 는 파이썬에서 완전히 동일하다.
    - 팀이 합의한 스타일을 따르는 것이 중요하다.

```python
"hello"
'hello'
```

- 문자열 내부에 `'` 또는 `"` 가 포함될 경우는 반대쪽 따옴표를 쓴다.

```python
"He said 'hello' to me"
'She said "yes" to the proposal'

# 백슬래시는 문자열을 더 복잡하고 읽기 어렵게 만든다.(권장하지 않음)
'He said \"hello\" to me'
```

- 트리플 쿼트 문자열(`""" """`)은 항상 이중 따옴표 사용한다.
    - PEP 257(파이썬 docstring 표준) 권장 사항이다.

```python
# ✔ 권장
"""Long string or docstring"""

# ❌ 권장하지 않음
'''Long string'''
```

- 문자열을 표현할 때 이런 규칙이 필요한 이유
    - 문자열 escaping 최소화 => 가독성 향상
    - 팀/프로젝트 전체에서 문자열 스타일 통일
    - docstring에서 삼중 큰 따옴표 `"""` 를 쓰는 것은 Python 공식 스타일
- 실무 팁!
    - 일반 문자열: `"   "`(Black 스타일)
    - 문자열 내부에 `"` 가 자주 등장하는 경우: `'   '`
    - docstring: `"""   """`(PEP 257 스타일)
    - 멀티라인 설명/SQL 쿼리/템플릿: `"""   """`
    - f-string: `f"..."` 또는 `f'''...'''` (단 escaping을 피하는 방안으로 선택) 

## 표현식과 문장에서 공백 사용법(Whitespace in Expressions and Statements)

### 불필요한 공백(Pet Peeves)

- 여는 괄호 직후 / 닫는 괄호 직전의 공백 금지

```python
# ✔ 권장
spam(ham[1], {eggs: 2})

# ❌ 잘못된 예
spam( ham[ 1 ], { eggs: 2 } )
```

- 쉼표 뒤 닫는 괄호 직전의 공백 금지

```python
# ✔ 권장
foo = (0,)

# ❌ 잘못된 예
bar = (0, ) # 괄호 앞 공백은 의미 없는 시각적 잡음
```

- 구두점(콤마, 세미콜론, 콜론) 앞의 공백 금지

```python
# ✔ 권장
if x == 4: print(x, y); x, y = y, x

# ❌ 잘못된 예
# - 구두점은 앞에 붙는 토큰이며, 뒤에 공백이 오는 것이 낫다.
if x == 4 : print(x , y) ; x , y = y , x 
```

- 슬라이싱에서의 콜론(:) 공백 규칙

```python
# ✔ 권장
# - 슬라이싱에서의 콜론(:)은 이항 연산자처럼 취급해야 한다.
# - 즉, 양쪽에 동일한 공백 규칙을 적용한다.
ham[1:9]
ham[1:9:3]
ham[:9:3] # 생략 자리에는 공백 사용하지 않음
ham[1::3]
ham[1:9:]
ham[lower:upper]
ham[lower:upper:]
ham[lower::step]
ham[lower + offset : upper + offset] # 연산자와의 일관성 유지
ham[: upper_fn(x) : step_fn(x)]
ham[:: step_fn(x)]

# ❌ 잘못된 예
# - 양쪽 공백이 불균형하다.
ham[lower + offset:upper + offset]
ham[1: 9]
ham[1 :9]
ham[1:9 :3]
ham[lower : : step]
ham[ : upper]
```

- 함수 호출의 여는 괄호 앞 공백 금지

```python
# ✔ 권장
spam(1)

# ❌ 잘못된 예
# - 함수 호출의 여는 괄호 앞 공백 금지
spam (1)  # 함수 “호출(call)”로 보이지 않게 만든다.
```

- 인덱싱/슬라이싱의 여는 괄호 앞 공백 금지

```python
# ✔ 권장
dct['key'] = lst[index]

# ❌ 잘못된 예
dct ['key'] = lst [index]  # 시각적으로 “연산”처럼 보이게 하는 오류
```

- 변수 정렬을 위한 “여러 개의 공백” 금지

```python
# ✔ 권장
x = 1
y = 2
long_variable = 3

# ❌ 잘못된 예
# - 변수 길이가 바뀌면 다시 맞춰야 함. 비효율!
# - diff 가 지저분해짐
x             = 1
y             = 2
long_variable = 3
```

    - membership / identity: `in`, `not in`, `is`, `is not`
    - Boolean: `and`, `or`, `not`

## 후행 콤마를 사용할 때(When to Use Trailing Commas)

후행 콤마(trailing comma)는 선택사항이지만 다음의 경우 사용하는 것이 좋다.

- 튜플에서 한 개짜리 요소(singleton tuple) 은 반드시 후행 쉼표 필요
    - 예) `('setup.cfg',)`    # 1개의 요소를 가진 튜플
    - 예) `('setup.cfg')`     # 그냥 문자열 

```python
# ✔ 권장
FILES = ('setup.cfg',)

# ❌ 잘못된 예
FILES = 'setup.cfg',  # 문법은 맞지만 가독성 나쁨.
```
