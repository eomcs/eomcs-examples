# 이름 짓는 규칙(Naming Conventions)

파이썬에서 권장하는 명명 규칙은 다음과 같다. 새로운 모듈과 패키지는 이러한 표준에 따라 작성한다. 
다만, 기존 라이브러리에서 사용하고 있는 스타일이 있다면 일관성을 위해 기존 스타일을 따르는 게 낫다.

## 최우선 원칙(Overriding Principle)

파이썬에서 이름을 짓는 모든 규칙의 최상위 기준은,

> 사용자에게 공개되는 API의 이름은 "구현 방식(implementation)"이 아니라 "사용 방식(usage)"을 반영해야 한다.

즉, 내부 구현이 어떻게 되어 있느냐는 중요하지 않고, 사용자가 이 기능을 어떻게 쓰게 될지를 기준으로 이름을 지어야 한다.

즉 다음 사항을 고려하라.

- 사용자가 쉽게 이해할 수 있어야 한다
  - 함수/클래스의 목적이 이름만 봐도 드러나야 한다.
- 사용자의 기대를 충족해야 한다
  - 기능은 이름이 암시하는 역할과 일치해야 한다.
  - 예: get_user() 라면 정말 가져오기(get) 역할만 해야지, 내부적으로 캐시를 초기화하거나 DB를 재구성하면 안 됨.
- 내부 구현 방식이 바뀌어도 이름은 유지되어야 한다
  - API의 이름이 구현 세부사항에 따라 바뀌면 사용자가 혼란스러워지기 때문.

구현이 아닌 사용성을 반영해야 한다.

```python
# ❌ 나쁜 예
def _calculate_value_from_cache_or_db():
  ...

# ✔ 권장
# - 내부 구현이 캐시인지 DB인지 드러나지 않는 이름이 더 적합하다.
# - 캐시를 쓰는지, DB를 쓰는지, 나중에 Redis로 바뀌는지는 사용자가 신경 쓸 필요가 없다.
# - 중요한 것은 “이 함수를 호출하면 값을 얻을 수 있다”는 사용성이다.
def get_value(user_id):
  ...
```

- 내부용(private) 함수라면, 내부 구현을 드러내도 괜찮다.

```python
# 이름 짓는 관례상 내부용 함수는 밑줄(_)로 시작한다.
def _load_value_from_cache():
    ...
```

- 정리
    - 공개용 API: 사용 방식을 반영하는 이름
    - 내부용 API: 구현 방식을 반영하는 이름도 허용
    - 기존 코드와 충돌하면: 일관성을 위해 기존 스타일을 따름

## 명명 스타일(Naming Styles)

- **b**: 소문자 한 글자

- **B**: 대문자 한 글자

- **lowercase**

- **lower_case_with_underscores**

- **UPPERCASE**

- **UPPER_CASE_WITH_UNDERSCORES**

- **CapitalizedWords**(= CapWords = CamelCase = StudlyCaps)
  - CapWords에서 약어는 모두 대문자:
    - `HTTPServerError`(O)
    - `HttpServerError`(X)

- **mixedCase**: 첫 글자가 소문자

- **Capitalized_Words_With_Underscores**: **비추!**

- **st_mode**, **st_size**, **st_mtime**: 고유 접두사를 사용(파이썬에서 많이 사용하지 않음)

- **_single_leading_underscore**: 내부용으로 사용되는 변수 또는 함수를 표시할 때.
  - 외부에서 직접 접근하거나 사용하지 말 것을 권고하는 암묵적인 규칙.
  - 문법적으로 사용하지 못하도록 강제하지는 않는다.
  - `from M import *` 는 `_`로 시작하는 객체는 가져오지 않는다.
  - `import M`을 사용하면 `_`로 시작하는 변수나 함수를 접근할 수 있다.

- **single_trailing_underscore_**: 파이썬 키워드와의 충돌을 피하기 위해 관례적으로 사용.
  - 예: `tkinter.Toplevel(master, class_='ClassName')`

- **__double_leading_underscore**: 클래스 속성의 이름
  - 서브 클래스와의 이름 충돌을 방지하기 위해, 이름 변형(name mangling)이 발생한다.
  - 예: FooBar 클래스의 `__foo` 인스턴스 변수가 있다면,
    - 클래스 내부: `__foo`
    - 클래스 외부: `FooBar.__foo`(X), `FooBar._FooBar__foo` (O)
      - 변수명 앞에 `_클래스명`이 붙는다.
  
- `__double_leading_and_trailing_underscore__`: "magic objects" 또는 "magic attributes" 에서 사용.
  - 파이썬 내부에서 사용하는 특수 예약 이름이다. 
  - 사용자가 마음대로 새로 만들어내면 안되고, 파이썬 문서에 있는 이름만 사용해야 한다.
  - 파이썬에서 특별한 동작을 하도록 설계된 객체나 속성:
    - `__init__`: 생성자
    - `__str__`: 문자열 표현
    - `__repr__`: 개발자용 문자열 표현
    - `__import__`: import 동작을 담당하는 내부 함수
    - `__file__`: 현재 실행 중인 파일 경로
    - `__name__`: 모듈 이름

### 피해야 할 문자(Names to Avoid)

- 숫자 0, 1과 혼동할 수 있는 문자는 피하라.

```python
# 숫자 0, 1과 혼동할 수 있는 문자는 피하라.
l   (소문자 L)
O   (대문자 O)
I   (대문자 I)

# 소문자 L 인지 숫자 1인지 구분하기 어렵다.
for l in list_of_items:
    ...
```

- 실무에서 많이 쓰는 단일 문자 변수
    - **i, j, k** → 반복문 인덱스(전통적인 수학 표기)
    - **x, y, z** → 좌표, 벡터 요소
    - **n** → count, size
    - **v / u / t**  → 벡터/시간 등 물리/수학적 의미

### ASCII 호환성

- 표준 라이브러리에서 사용되는 식별자는 PEP 3131 의 정책 섹션에 설명된 대로 ASCII 호환이어야 한다.
- 즉 표준 라이브러리의 모든 식별자는 ASCII 문자로 작성하라.
  - 예: 변수명, 함수명, 클래스명, 메서드명, 모듈명, 패키지명, 속성명 등 
- 리터럴(literal)과 주석도 ACSII로 작성하라.
- 단, 작성자 이름은 ASCII가 아닌 문자로 작성하더라도 읽을 수 있도록 라틴 문자로 표시해 줘야 한다.

```python
# ✔ 권장:
message = "안녕하세요"     # 문자열은 자유롭게 non-ASCII 가능

# ❌ 나쁜 예:
사용자 = "홍길동"          # 코드 에디터, 도구, 협업에서 문제 발생 가능

# ✔ 권장:
def calculate(x, y):
    return x + y

# ❌ 나쁜 예:
def 계산하기(x, y):
    return x + y

# ✔ 권장:
message = "사용자 이름을 입력하세요."
```

```python
# ✔ 권장:
# - import 문에서 이름이 길면 읽기 불편 해진다.
# - 명확하고 일관된 네이밍이 유지보수성 향상에 도움.
# - Python 표준 라이브러리도 대부분 이런 패턴을 따름.
utils.py
parser.py
string_tools.py
data_loader.py

# ❌ 나쁜 예:
StringTools.py     # CamelCase 금지
DataLoader.PY      # 대문자 포함 금지
MyModule.py        # CamelCase 금지
```

- C/C++ 확장 모듈과 매칭되는 Python 모듈 네이밍 규칙
    - C/C++로 구현된 저수준 모듈 이름은 앞에 언더스코어(`_`)를 붙인다.
    - 이 모듈을 감싸는 고수준 Python 모듈은 일반 이름을 사용한다.
    - 예: 표준 라이브러리

```python
# ✔ 권장:
_parser.cpython-38-x86_64-linux-gnu.so   → _parser
parser.py                                → public interface
```

- 패키지 + 모듈 이름 규칙 예시

```python
# ✔ 권장:
mypackage/
    __init__.py
    http_client.py
    parser.py
    file_loader.py

# ❌ 나쁜 예:
MyPackage/              # 대문자 사용
    HttpTools.py        # CamelCase
    File-Loader.py      # 하이픈 금지
    xmlParser.py        # CamelCase + 혼합 스타일
```

### 타입 변수(PEP 484) 이름(Type Variable Names)

- CapWords(PascalCase)를 사용
    - 클래스 이름과 동일한 규칙 적용

```python
# ✔ 권장:
T = TypeVar("T")
KT = TypeVar("KT")
VT = TypeVar("VT")

# ❌ 나쁜 예:
value_type = TypeVar("value_type")
t = TypeVar("t")
number = TypeVar("number")
value = TypeVar("value")
```

- 공변성(covariant), 반공변성(contravariant) TypeVar 네이밍 규칙
    - 공변성: 타입명 + `_co`
    - 반공변성: 타입명 + `_contra`

```python
# ✔ 권장:
VT_co = TypeVar('VT_co', covariant=True)
KT_contra = TypeVar('KT_contra', contravariant=True)
```

```python
# Covariant 예제
from typing import Generic, TypeVar

T_co = TypeVar("T_co", covariant=True)

class ReadOnlyList(Generic[T_co]):
    def __init__(self, items: list[T_co]):
        self._items = items

    def get(self, index: int) -> T_co:
        return self._items[index]
```

```python
# Contravariant 예제
from typing import Generic, TypeVar

T_contra = TypeVar("T_contra", contravariant=True)

class Comparator(Generic[T_contra]):
    def compare(self, a: T_contra, b: T_contra) -> int:
        ...
```

### 예외 이름(Exception Names)

- 예외는 클래스이므로 클래스 네이밍 규칙(CapWords) 사용
    - 이름 끝에 "Error"를 붙인다.

```python
# ✔ 권장:
class InvalidTokenError(Exception):
    pass

class DatabaseConnectionError(Exception):
    pass

class PermissionDeniedError(Exception):
    pass
```

- 실무에서의 예외 설계 규칙

```python
# 규칙 1: 모든 커스텀 예외는 Exception 또는 RuntimeError 기반이어야 한다.
class AppError(Exception):
    """Base class for all app errors."""

# 규칙 2: 도메인별 에러 그룹을 만들고 Error로 끝나는 이름 사용
class AuthError(AppError):
    pass

class InvalidCredentialsError(AuthError):
    pass

class TokenExpiredError(AuthError):
    pass

# 규칙 3: Error가 아닌 “제어 흐름용 예외”는 Error 없이 이름 지정
class StopIteration(Exception):
    pass

class Cancelled(Exception):
    pass

class Abort(Exception):
    pass

class Timeout(Exception):
    pass
```

### 전역 변수 이름(Global Variable Names)

- 전역 변수는 기본적으로 함수 이름 규칙(snake_case)을 따른다.
    - 함수명/변수명과 동일하게 **lowercase + underscores** 형식을 사용
    - 전역 변수는 **한 모듈 내부에서만 사용**되는 것이 바람직하다.
    - 전역 변수는 **다른 모듈에서 직접 접근하지 않는 것**이 좋다.

```python
# ✔ 권장:
global_counter
config_path
default_timeout
```

- 모듈이 `from module import *` 형태로 사용될 수 있는 경우,
    - `__all__`을 사용해 외부로 공개할 이름을 통제하라

```python
# ✔ 권장:
# - 공개되면 안 되는 전역 변수는 __all__에 넣지 말라.
__all__ = ["public_function", "MyClass"]
```

- 오래된 관례로 언더스코어(_) 접두사를 사용할 수 있음
    - 언더스코어로 시작하면 PEP 8에 따라 "**모듈 내부용**(module non-public)"이라는 의미
    - `from module import *` 했을 때도 자동으로 import되지 않는다.

```python
# ✔ 권장:
_internal_cache = {}
```

- 실무에서 자주 쓰는 전역 변수 스타일

```python
# 상수 값(global constant)은 ALL_CAPS
MAX_RETRY = 3
DEFAULT_TIMEOUT = 5.0
API_VERSION = "1.0"

# 내부 전역 변수는 _name 으로 표시
_internal_state = {}
_current_user = None

# 공개 API 전역 변수는 __all__로 명확히 노출
__all__ = ["PI", "calculate"]

PI = 3.14159

def calculate(): ...
```

- 전역 상태를 가능하면 피하고, class / function scope로 둘 것을 권장

### 함수 이름과 변수 이름(Function and Variable Names)

- 함수(Function) 이름 규칙
    - 함수 이름은 **소문자 + 밑줄** 형식 사용
    - 함수 이름은 동사(Verb) 또는 동사구(Verb Phrase)로 표현

```python
# ✔ 권장:
def load_data(): ...
def get_user_info(): ...
def calculate_total_price(): ...

# ❌ 나쁜 예:
def LoadData(): ...       # CamelCase 금지
def getUserInfo(): ...    # mixedCase 금지
def calculateTotal(): ... # mixedCase 금지
```

- 변수(variable) 이름도 함수 이름과 동일한 규칙
    - 변수 이름은 **소문자 + 밑줄** 형식 사용
    - 변수 이름은 명사(Noun) 또는 명사구(Noun Phrase)로 표현

```python
# ✔ 권장:
user_count = 0
total_price = 100
max_length = 50

# ❌ 나쁜 예:
UserCount = 0           # 클래스 이름처럼 보임
totalPrice = 100        # mixedCase 지양
maxLength = 50
```

- 실무:
    - 기존 코드가 mixedCase라면, 기존 스타일과의 호환을 위해 mixedCase 유지
    - 새 코드에서는 절대 mixedCase를 사용하지 않음 (레거시 호환 목적 제외)

### 함수와 메서드 아규먼트(Function and Method Arguments)

- 인스턴스 메서드의 첫 번째 인자: `self`
    - 인스턴스 메서드의 첫 번째 인자는 항상 `self`로 명명
    - 다른 이름을 사용해도 문법적으로 문제없지만, 권장하지 않음

```python
# ✔ 권장:
class User:
    def save(self):
        print(self.name)


# ❌ 나쁜 예:
class User:
    def save(this):   # 파이썬 커뮤니티에서는 매우 나쁜 스타일
        print(this.name)    
```

- 클래스 메서드의 첫 인자: `cls`
    - 클래스 메서드의 첫 번째 인자는 항상 `cls`로 명명
    - 다른 이름을 사용해도 문법적으로 문제없지만, 권장하지 않음

- Name Mangling(이름 맹글링) 활성화: **__double_leading_underscore** 스타일 사용
    - 서브 클래스와의 이름 충돌을 피하기 위해 이름 맹글링 사용

```python
# 클래스 정의
class Foo:
    def __a(self):  # 내부적으로 _Foo__a 로 변환된다.
        print("Foo.__a() 호출됨!")

class ChildFoo(Foo):
    def __a(self):  # 내부적으로 _ChildFoo__a 로 변환된다.
        print("ChildFoo.__a() 호출됨!")

    def test(self):
        # 클래스 내부에서는 원래 이름으로 접근 가능
        self.__a()  # ChildFoo.__a() 호출됨!

        # mangling 된 이름으로도 접근 가능
        self._ChildFoo__a()  # ChildFoo.__a() 호출됨!

        # 부모 클래스의 메서드는 반드시 mangling 된 이름으로 접근해야 함
        self._Foo__a()  # 부모 클래스의 __a() 호출 가능    

# 클래스 사용
child_foo = ChildFoo()
child_foo.test()
print("--------------")

child_foo._ChildFoo__a()
child_foo._Foo__a()

# ❌ 나쁜 예:
child_foo.__a()  # AttributeError 발생 - 외부에서는 원래 이름으로 접근 불가
```

### 상속을 고려한 설계(Designing for Inheritance)   

- 상속을 염두에 둔 클래스의 “**속성(메서드 + 인스턴스 변수) 공개 범위**”를 명확히 결정하라
    - **공개(public)**: 
        - **외부 사용자에게 약속한 기능**.
        - 클래스 외부(서브 클래스나 외부 모듈)에서 자유롭게 접근 가능한 속성.
        - 클래스 사용자(client)가 이 속성을 안전하게 사용할 수 있다는 의미.
        - 이후 버전에서도 호환성을 유지하겠다는 개발자의 약속을 내포.
        - **호환성 보장**: 공개 속성은 변경하거나 제거하면 사용 중인 코드에서 문제가 발생할 수 있기 때문.
        - **문서화 대상**: 공개 속성은 API로 간주되어 용도와 사용법을 명확히 문서화 해야 한다.
    - **Subclass API(= protected)**: 
        - 상속하여 재정의하거나 확장할 때 사용하도록 허용하는 속성.
    - **비공개(private)**: 
        - **클래스 내부에서만 사용할 목적**으로 만든 속성이다.
        - 제3자(third parties; 클래스를 사용하는 외부 코드 또는 서브 클래스)에게 사용을 권장하지 않는다.
        - 필요에 따라 **언제든 변경되거나 삭제될 수 있음**을 암시한다.
        - **호환성 미보장**: 외부에서 사용하지 않을 것을 가정하기 때문에 언제든 변경하거나 삭제할 수 있다.
        - **문서화 대상이 아님**: API 문서에서 제외된다. 내부 구현 세부 사항으로 간주된다.
        - **관례적으로 보호 대상**: 사용을 막을 수는 없지만, 관례적으로 외부 사용을 방지한다.
        - `private` 이라는 용어는 사용하지 않는다. Python에 `private` 문법이 없다.
    - 결정 기준:
        - 모르겠으면 일단 비공개 하라.
        - 나중에 공개 속성을 비공개로 만드는 것보다 비공개를 공개로 만드는 것이 더 쉽다.

- Public vs Non-public 기본 규칙
    - 공개(public) 속성: 이름에 밑줄(`_`)이 없음.
    - 비공개(non-public) 속성: 이름에 밑줄(`_`)이 있음.

```python
# Public 속성 → 언더스코어로 시작하지 않는다
# - 공개 속성에는 앞에 밑줄이 없어야 한다.
# - 이름이 그대로 공개 API임을 의미
# - 하위 호환 유지에 책임이 생김
class User:
    name = ""
    def save(self): ...

# Non-public 속성 → 한 개의 언더스코어 _name
# - 외부에서 사용하지 않음을 나타내는 컨벤션이다.
# - import * 시 노출되지 않는다.
# - 하지만 진정한 “private”은 아니다.(파이썬에서는 “private” 개념이 없다)
# - 즉, 언제든 값을 변경 할 수 있다.
class User:
    _cache = {}
    def _load(self): ...
```

- 파이썬은 왜 “private”이라는 말을 쓰지 않는가?
    - “정보 은닉” 그 자체는 Python의 철학이 아니다.
    - 파이썬의 객체지향 철학은 다른 언어(C++, Java)와 근본적으로 다르다.
    - **C++**, **Java** 같은 언어에서 private은 **강제된 인닉(enforced hiding)을 의미**한다.
    - **Python**은 강제 대신 “**합의에 기반한 구조**”일 뿐이다.
    - 즉, **"이건 건드리지 않는게 좋아요"라고 합의**한 비강제적 신사협정이다.

- 서브 클래스를 고려한 클래스의 **Subclass API(Protected 역할) 설계 기준**:
    - **어떤 메서드를 오버라이드해도 되는지**
    - **어떤 속성은 서브 클래스가 사용하면 안 되는지**

- 공개 속성 이름이 “키워드(keyword)”와 충돌한다면 → 뒤에 `_`
    - 함수 아규먼트 며명 규칙과 동일하다.

- 공개 데이터 속성은 단순하게 노출하라. 
    - 간단한 공개 데이터 속성의 경우 **속성 이름만 노출**하는 것이 가장 좋다.
    - 추후 **accessor/mutator 필요하면 데코레이터를 사용**하여 속성을 메서드로 전환할 수 있다.
        - accessor(getter): `@property`
        - mutator(setter): `@프로퍼티명.setter`
        - 기존의 접근 방식(속성 이름)을 유지하면서 기능을 확장할 수 있다.
        - 이는 유연성과 확장성을 높이는 설계 방식이다.

- accessor/mutator 또는 @property **데코레이터를 사용할 때 주의점**!
    - 기능적 동작에는 부작용(side-effects)이 없어야 한다.
    - 즉 속성 접근이나 변경할 때 외부 자원을 변경한다거나 파일 쓰기 등을 해서는 안된다.
    - 일시적으로 속성 값을 보관하는 캐싱(허용 가능한 부작용)은 상관없다.
    - 속성의 설계 목표는 단순하고 직관적인 사용과 예측 가능성을 유지하는 것이다.
    - **연산 비용이 큰 경우 속성을 @property를 사용하지 말라.**
        - @property 데코레이터는 속성처럼 보이지만 실제 함수이다.
        - 따라서 @property를 사용하는 caller 입장에서는 접근이 간단하고 빠른 동작일 것이라 가정한다.
        - 그런데 호출자의 기대와 달리 실제 동작 간에 불일치가 발생하여 성능이나 예기치 않은 문제가 발생할 수 있다.
        - 명시적으로 메서드를 호출하는 방식으로 설계하라.

    - 서브 클래스와의 이름 충돌을 방지하기 위해 **이름 맹글링(name mangling) 사용**한다.
    - 단 디버깅이나 `__getattr__()`를 사용할 때 불편할 수 있다.
    - 모든 사람이 **name mangling**을 좋아하는 것은 아니다.
    - 실수로 이름이 충돌하는 것을 피해야 하는 필요성과 고급 호출자의 잠재적 사용 사이에서 균형을 고려하라.

## 공개 및 내부 인터페이스(Public and Internal Interfaces)

- 백워드 호환(backwards compatibility) 보장은 오직 public API에만 적용된다.
- 그러므로 사용자(개발자)가 public과 internal API를 분명히 구분할 수 있도록 설계해야 한다.

즉, 어떤 함수/클래스/모듈이 “공식적인(public) 인터페이스인지”를
명확히 알려야 하고, 나머지는 모두 internal로 간주해야 한다.

- 무엇이 public 인가?
    - 문서(documentation)에 포함된 인터페이스는 public 이다.
        - 단, 문서에서 “internal”, “provisional”로 명시하지 않았을 경우.
        - 문서에 없는 인터페이스는 internal로 간주한다.
        - 심지어 소스 코드상 public 처럼 보이더라도 문서에 없다면 public이 아니다.
    - 이름 앞에 `_`가 붙지 않았고,
    - `__all__`에 포함되어 있고,(선택적이지만 권장)
    - 상위 네임스페이스(패키지, 모듈, 클래스)가 public 이고,
    - 다른 파일에 직접 import 해도 괜찮도록 설계된 것.

- Public API를 명확하게 하기 위한 도구: `__all__`
    - 모듈은 `__all__`로 자신의 public API를 명시해야 한다.
    - `__all__ = []`이라면 “이 모듈에는 public API가 없음”을 의미한다.

```python
# foo.py
__all__ = ["Parser", "load"]

class Parser: ...   # public
def load(): ...     # public
def _helper(): ...  # internal
```

```python
# 사용자 코드
from foo import *  # Parser, load 만 import 된다.
```

- Internal API는 반드시 언더스코어(`_`)로 표시해야 한다
    - 문서에 없고,
    - 이름에 `_`가 붙어 있고
    - `__all__`에 포함되지 않은 것

```python
def _internal_helper(): ...
class _InternalClass: ...
```

- 포함하는 네임스페이스(패키지, 모듈, 클래스)가 internal 이면, 그 안의 모든 것도 internal 이다.

package/
    _internal/
        helper.py

```python
# 사용자 코드
# - 다음과 같은 사용은 권장되지 않는다.
from package._internal.helper import do_something
```

- 다른 모듈에서 import한 이름은 그 모듈의 공식 public API가 아니다.
    - **import 한 이름은 언제든 바꿀 수 있는 내부 구현일 뿐이니, 외부 모듈이 그걸 믿고 쓰면 안된다.**

```python
# api.py
import json

# json 모듈은 다음과 같이 내부 구현에 사용하기 위해 import 한 것이다.
# api 모듈의 공식 API가 아니다.(implementation detail)
def load_config(path):
    with open(path) as f:
        return json.load(f)
```

```python
# other.py
from api import json  # ❌ api 모듈을 통한 우회 import 금지

# - api는 json을 공식 public API로 노출하지 않았다.
# - 만약 api.py 에서 json 대신 orjson을 쓰도록 변경한다면, 이 코드는 깨지게 된다.
```

- 패키지의 `__init__.py`에서 **의도적으로 하위 모듈을 export 하는 경우**에도 간접 임포트가 허용된다.

```python
mypkg/
    __init__.py
    models.py
    services.py
```
